# Exercice 01

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_iris 
iris = load_iris()
X = iris.data 
y= iris.target 
```

```python
kmean = KMeans(n_clusters = 3)
```

```python
km = kmean.fit(X)
km
```

```python
km.labels_
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, km.labels_)
print(cm)
```

```python
sns.set(style ="whitegrid")
ax = sns.swarmplot(km.labels_, X[:, 0])
ax.set(xlabel ='Clusters', ylabel ='sepal_length')
plt.show()
```

```python
ax = sns.swarmplot(km.labels_, X[:, 1])
ax.set(xlabel ='Clusters', ylabel ='sepal_width')
plt.show()
```

```python
ax = sns.swarmplot(km.labels_, X[:, 2])
ax.set(xlabel ='Km.labels', ylabel ='Patel_length')
plt.show()
```

# Exercice 05 with K-means

```python
import pandas as pd 
Mall = pd.read_csv('Mall_Customers.csv')
```

```python
Mall
```

```python
X = Mall.iloc[:,[3,4]].values
```

here, I used Elbow method  

```python
withness = []
for i in range(1,10):
    kmean = KMeans(n_clusters = i, random_state=0)
    kmean.fit(X)
    withness.append(kmean.inertia_)

plt.plot(range(1,10),withness)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

So, what i saw the results of the Elbow method , I conclude the number optimal of clusters is equal = 5

```python
kmean = KMeans(n_clusters = 5, random_state=0)
kmean.fit(X)
plt.scatter(X[:,0], X[:,1], c=kmean.labels_)
plt.xlabel('Annual')
plt.ylabel('score')
plt.show()
```
